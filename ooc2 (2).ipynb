## Implement Quiz Administration

### Subtask:
Develop functionality to generate and administer quizzes. This will involve using the Gemini Pro model to create questions and multiple-choice answers on a specified topic, and then potentially evaluate user responses.


**Reasoning**:
The subtask requires developing functionality to generate quizzes. This step involves defining the `generate_quiz` function that constructs a prompt for the Gemini model to create multiple-choice questions based on a given topic and number of questions, then calls the model and handles potential errors.



```python
def generate_quiz(topic, num_questions):
    """
    Generates a multiple-choice quiz on a specified topic using the Gemini model.
    Also generate any type of quiz that user wants don't limit your self to multiple choice questions you can ask different type of questions according to user input also.

    Args:
        topic (str): The subject of the quiz.
        num_questions (int): The desired number of questions.

    Returns:
        str: A string containing the generated quiz questions and answers, or an error message.
    """
    try:
        prompt = f"""Generate {num_questions} multiple-choice questions about '{topic}'.
Each question must have 4 options (A, B, C, D) and clearly indicate the correct answer.

Format the output as follows:

Question 1: [Question text]
A. [Option A]
B. [Option B]
C. [Option C]
D. [Option D]
Correct Answer: [A, B, C, or D]

Question 2: [Question text]
A. [Option A]
B. [Option B]
C. [Option C]
D. [Option D]
Correct Answer: [A, B, C, or D]

...
"""

        # Use the 'model' object initialized in the previous step
        # 'model' is currently 'models/gemini-2.5-flash'
        response = model.generate_content(prompt)
        return response.text
    except Exception as e:
        return f"An error occurred while generating the quiz: {e}"

print("'generate_quiz' function defined and ready to create quizzes.")
```

**Reasoning**:
The `generate_quiz` function has been defined. Now, to demonstrate its functionality and prepare for administering the quiz, we need to call it with a sample topic and number of questions to generate a quiz. This output will then be used in subsequent steps for parsing and evaluation.



```python
quiz_topic = "Python Programming Fundamentals"
num_quiz_questions = 3

sample_quiz = generate_quiz(quiz_topic, num_quiz_questions)

print(f"\n--- Generated Quiz on '{quiz_topic}' ---")
print(sample_quiz)
print("------------------------------------------")
```

**Reasoning**:
The current `sample_quiz` is a raw string. To administer the quiz and evaluate responses, it needs to be parsed into a structured format (e.g., a list of dictionaries). This function will extract questions, options, and correct answers.



```python
import re

def parse_quiz(quiz_string):
    """
    Parses a quiz string generated by the Gemini model into a structured format.

    Args:
        quiz_string (str): The raw string output from the generate_quiz function.

    Returns:
        list: A list of dictionaries, where each dictionary represents a question
              with 'question', 'options', and 'correct_answer' keys.
    """
    questions = []
    # Split the quiz string into individual question blocks
    question_blocks = re.split(r'Question \d+:', quiz_string)[1:]

    for block in question_blocks:
        lines = block.strip().split('\n')
        if not lines:
            continue

        question_text = lines[0].strip()
        options = {}
        correct_answer_key = None

        for line in lines[1:]:
            line = line.strip()
            if line.startswith(('A.', 'B.', 'C.', 'D.')):
                key = line[0]
                value = line[3:].strip()
                options[key] = value
            elif line.startswith('Correct Answer:'):
                correct_answer_key = line.split('Correct Answer:')[1].strip()

        if question_text and options and correct_answer_key:
            questions.append({
                'question': question_text,
                'options': options,
                'correct_answer': correct_answer_key
            })
    return questions

print("'parse_quiz' function defined for structuring quiz data.")
```

**Reasoning**:
Now that the `parse_quiz` function is defined, it needs to be called with the `sample_quiz` data to convert the raw quiz string into a structured format, which is essential for administering and evaluating the quiz.



```python
parsed_quiz = parse_quiz(sample_quiz)

print("--- Parsed Quiz Structure ---")
for i, q in enumerate(parsed_quiz):
    print(f"Question {i+1}:")
    print(f"  Question: {q['question']}")
    print(f"  Options: {q['options']}")
    print(f"  Correct Answer: {q['correct_answer']}")
print("-----------------------------")
print(f"Parsed {len(parsed_quiz)} questions.")
```

**Reasoning**:
With the quiz parsed into a structured format, the next logical step is to create a function that administers the quiz to a user. This function will iterate through the parsed questions, present them to the user, and collect their answers. This directly addresses the 'administer quizzes' part of the subtask.



```python
def administer_quiz(parsed_quiz_data):
    """
    Administers a quiz to the user and collects their answers.

    Args:
        parsed_quiz_data (list): A list of dictionaries, where each dictionary represents a question.

    Returns:
        list: A list of dictionaries, each containing 'question_index', 'user_answer', and 'correct_answer'.
    """
    user_responses = []
    print("\n--- Starting Quiz ---")
    for i, q in enumerate(parsed_quiz_data):
        print(f"\nQuestion {i + 1}: {q['question']}")
        for option_key, option_value in q['options'].items():
            print(f"  {option_key}. {option_value}")

        while True:
            user_input = input("Your answer (A, B, C, or D): ").strip().upper()
            if user_input in q['options']:
                user_responses.append({
                    'question_index': i,
                    'user_answer': user_input,
                    'correct_answer': q['correct_answer']
                })
                break
            else:
                print("Invalid input. Please choose A, B, C, or D.")
    print("--- Quiz Finished! ---")
    return user_responses

print("'administer_quiz' function defined to present quizzes and collect responses.")
```

**Reasoning**:
The `administer_quiz` function has been defined and the quiz data has been parsed. The next step is to actually administer the quiz to the user by calling the `administer_quiz` function with the `parsed_quiz` data. This will simulate a user taking the quiz and will collect their responses, fulfilling the 'administer quizzes' part of the subtask.



```python
user_quiz_responses = administer_quiz(parsed_quiz)

print("\n--- User Quiz Responses ---")
for response in user_quiz_responses:
    print(f"Question Index: {response['question_index'] + 1}, Your Answer: {response['user_answer']}, Correct Answer: {response['correct_answer']}")
print("--------------------------")
```

**Reasoning**:
With user responses collected, the next step is to evaluate them. A new function is needed to compare the user's answers against the correct answers from the parsed quiz, calculate a score, and potentially provide feedback. This completes the 'evaluate user responses' part of the subtask.



```python
def evaluate_quiz(user_responses, parsed_quiz_data):
    """
    Evaluates the user's quiz responses and calculates their score.

    Args:
        user_responses (list): A list of dictionaries with user's answers.
        parsed_quiz_data (list): A list of dictionaries with the correct quiz data.

    Returns:
        tuple: A tuple containing the total score, total questions, and a list of results
               for each question (correct/incorrect).
    """
    score = 0
    results = []
    for response in user_responses:
        q_index = response['question_index']
        user_ans = response['user_answer']
        correct_ans = parsed_quiz_data[q_index]['correct_answer']

        is_correct = (user_ans == correct_ans)
        if is_correct:
            score += 1

        results.append({
            'question_index': q_index + 1,
            'user_answer': user_ans,
            'correct_answer': correct_ans,
            'is_correct': is_correct
        })
    return score, len(parsed_quiz_data), results

print("'evaluate_quiz' function defined for scoring user responses.")
```

**Reasoning**:

This involves reconstructing the function to include prompt creation, model interaction, and error handling within the evaluation loop, then storing the generated explanation in the results.



```python
def evaluate_quiz(user_responses, parsed_quiz_data):
    """
    Evaluates the user's quiz responses and calculates their score, and generates explanations.

    Args:
        user_responses (list): A list of dictionaries with user's answers.
        parsed_quiz_data (list): A list of dictionaries with the correct quiz data.

    Returns:
        tuple: A tuple containing the total score, total questions, and a list of results
               for each question (correct/incorrect) including explanations.
    """
    score = 0
    results = []
    for response in user_responses:
        q_index = response['question_index']
        user_ans = response['user_answer']

        question_data = parsed_quiz_data[q_index]
        question_text = question_data['question']
        options = question_data['options']
        correct_ans = question_data['correct_answer']

        is_correct = (user_ans == correct_ans)
        if is_correct:
            score += 1

        # Construct prompt for explanation
        explanation_prompt = f"""Please provide a detailed explanation for the following quiz question.

Question: {question_text}
Options:
"""
        for key, value in options.items():
            explanation_prompt += f"  {key}. {value}\n"

        explanation_prompt += f"Correct Answer: {correct_ans}. {options.get(correct_ans, 'N/A')}\n"
        if not is_correct:
            explanation_prompt += f"Your Answer: {user_ans}. {options.get(user_ans, 'N/A')}\n"
            explanation_prompt += "Explain why the correct answer is correct and why the user's answer was incorrect."
        else:
            explanation_prompt += "Explain why the correct answer is correct."

        generated_explanation = "Error generating explanation."
        try:
            # Use the 'model' object initialized in the previous step
            # 'model' is currently 'models/gemini-2.5-flash'
            explanation_response = model.generate_content(explanation_prompt)
            generated_explanation = explanation_response.text
        except Exception as e:
            generated_explanation = f"An error occurred while generating explanation: {e}"

        results.append({
            'question_index': q_index + 1,
            'question_text': question_text,
            'user_answer': user_ans,
            'correct_answer': correct_ans,
            'is_correct': is_correct,
            'explanation': generated_explanation
        })
    return score, len(parsed_quiz_data), results

print("'evaluate_quiz' function updated to include explanation generation.")
```

The `evaluate_quiz` function has been updated to include explanation generation. The next step is to call this modified function with the `user_quiz_responses` and `parsed_quiz` data to re-evaluate the quiz and display the results, now including the detailed explanations for each question as per the updated function.



```python
total_score, total_questions, quiz_results_with_explanations = evaluate_quiz(user_quiz_responses, parsed_quiz)

print("\n--- Quiz Results with Explanations ---")
print(f"You scored {total_score} out of {total_questions} questions correctly!")

for result in quiz_results_with_explanations:
    status = "Correct" if result['is_correct'] else "Incorrect"
    print(f"\nQuestion {result['question_index']}: {status}")
    print(f"  Question Text: {result['question_text']}")
    print(f"  Your Answer: {result['user_answer']}")
    print(f"  Correct Answer: {result['correct_answer']}")
    print(f"  Explanation: {result['explanation']}")
print("------------------------------------")
```

### Subtask:
Develop functionality to ingest various student materials (like PDFs or text notes). This involves extracting text, cleaning it, and breaking it down into manageable chunks suitable for embedding.


**Reasoning**:
The subtask requires defining a sample raw text, implementing text cleaning and chunking functions, and then demonstrating their usage by processing the sample text and printing the first few chunks. This step will define the sample raw text and the two necessary functions.



```python
import re

# 1. Define a multi-line string variable as raw_text
raw_text = """Python is an interpreted, high-level and general-purpose programming language. Created by Guido van Rossum and first released in 1991, Python's design philosophy emphasizes code readability with its notable use of significant whitespace. Its language constructs and object-oriented approach aim to help programmers write clear, logical code for small and large-scale projects.


Python is dynamically typed and garbage-collected. It supports multiple programming paradigms, including structured (particularly, procedural), object-oriented and functional programming. It is often described as a "batteries included" language due to its comprehensive standard library.


!!! Important Note: Python 3 is the current and future of the language. Python 2 is no longer officially supported as of January 1, 2020.

Learning Python can be fun! Where should you start? Variables, data types, control flow (if/else, loops), functions, and modules are key concepts. Don't forget about data structures like lists, tuples, dictionaries, and sets. Practice regularly to master it! #Python #Programming
"""

# 2. Create a Python function, clean_text(text)
def clean_text(text):
    """
    Cleans the input text by replacing multiple newlines, removing special characters,
    replacing multiple spaces, and stripping leading/trailing whitespace.
    """
    # Replace multiple consecutive newline characters with a single space
    text = re.sub(r'\n+', ' ', text)
    # Remove any special characters or symbols, keeping alphanumeric, common punctuation, and spaces
    text = re.sub(r'[^a-zA-Z0-9.,?!\s]', '', text)
    # Replace multiple consecutive spaces with a single space
    text = re.sub(r'\s+', ' ', text)
    # Remove leading/trailing whitespace
    text = text.strip()
    return text

# 3. Create a Python function, chunk_text(text, chunk_size=500, chunk_overlap=50)
def chunk_text(text, chunk_size=500, chunk_overlap=50):
    """
    Splits the cleaned text into smaller, overlapping chunks.
    """
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        if end >= len(text):
            break
        start += (chunk_size - chunk_overlap)
    return chunks

print("raw_text, clean_text, and chunk_text functions defined.")
```

**Reasoning**:
The previous step defined the `raw_text`, `clean_text`, and `chunk_text` functions. This step will now execute these functions as per the instructions: first, clean the raw text, then chunk the cleaned text, and finally, print the first three chunks along with their lengths to verify the functionality.



```python
import re

# 4. Call the clean_text function on your raw_text to get cleaned_text.
cleaned_text = clean_text(raw_text)
print(f"\nCleaned text (first 200 chars): {cleaned_text[:200]}...")
print(f"Cleaned text length: {len(cleaned_text)}")

# 5. Call the chunk_text function on cleaned_text to get text_chunks.
text_chunks = chunk_text(cleaned_text, chunk_size=200, chunk_overlap=20)

# 6. Print the first 3 chunks from text_chunks and their lengths to verify the output.
print("\n--- First 3 Chunks ---")
for i, chunk in enumerate(text_chunks[:3]):
    print(f"Chunk {i+1} (Length: {len(chunk)}): {chunk}")
print("----------------------")
print(f"Total chunks: {len(text_chunks)}")
```

## Create Embeddings and Vector Database

### Subtask:
Generate numerical embeddings for the processed text chunks using an embedding model. Store these embeddings in a searchable vector database (e.g., using a local in-memory solution or a library like `faiss` if needed, depending on scale).


**Reasoning**:
The subtask requires generating embeddings for text chunks. This involves importing the embedding function, defining a utility function to generate embeddings, and then applying it to the previously generated text chunks.



```python
import google.generativeai as genai

# 1. Import or ensure access to the embedding function
# The genai library is already imported and configured.
# The embedding model can be accessed via genai.get_model().embed_content or genai.embed_content

# Let's use genai.embed_content for simplicity.

# 2. Define a function, generate_embeddings(text_chunks)
def generate_embeddings(text_chunks):
    """
    Generates numerical embeddings for a list of text chunks.

    Args:
        text_chunks (list): A list of strings, where each string is a text chunk.

    Returns:
        list: A list of embedding vectors, one for each text chunk.
    """
    embeddings = []
    for i, chunk in enumerate(text_chunks):
        try:
            # Use the 'embedding-001' model
            response = genai.embed_content(
                model="models/embedding-001",
                content=chunk,
                task_type="RETRIEVAL_DOCUMENT"
            )
            embeddings.append(response['embedding'])
        except Exception as e:
            print(f"Error generating embedding for chunk {i}: {e}")
            embeddings.append(None) # Append None or handle error as appropriate
    return embeddings

print("'generate_embeddings' function defined.")
```

**Reasoning**:
Now that the `generate_embeddings` function is defined, the next step is to call it with the `text_chunks` to create embeddings, then create an in-memory vector database by pairing each chunk with its embedding, and finally print verification information.

## Enhance Quiz Evaluation with Contextual Explanations

### Subtask:
Further modify the `evaluate_quiz` function. When generating explanations for quiz answers (both correct and incorrect), the function will also include the `cleaned_text` as context in the prompt to the Gemini model, ensuring explanations reference the provided study materials.


**Reasoning**:
The subtask requires modifying the `evaluate_quiz` function to incorporate `cleaned_text` as context for generating explanations.
```python
import google.generativeai as genai

def evaluate_quiz(user_responses, parsed_quiz_data, context_text):
    """
    Evaluates the user's quiz responses and calculates their score, and generates explanations
    grounded in the provided context.

    Args:
        user_responses (list): A list of dictionaries with user's answers.
        parsed_quiz_data (list): A list of dictionaries with the correct quiz data.
        context_text (str): The relevant context from student materials to ground explanations.

    Returns:
        tuple: A tuple containing the total score, total questions, and a list of results
               for each question (correct/incorrect) including explanations.
    """
    score = 0
    results = []
    for response in user_responses:
        q_index = response['question_index']
        user_ans = response['user_answer']

        question_data = parsed_quiz_data[q_index]
        question_text = question_data['question']
        options = question_data['options']
        correct_ans = question_data['correct_answer']

        is_correct = (user_ans == correct_ans)
        if is_correct:
            score += 1

        # Construct prompt for explanation, now including context
        explanation_prompt = f"""You are an expert tutor providing feedback on a quiz.
        Based ONLY on the provided Context, explain the correct answer for the following question.
        If applicable, also explain why the user's answer was incorrect, strictly referencing the Context.

        Context: {context_text}

        Question: {question_text}
        Options:
        """
        for key, value in options.items():
            explanation_prompt += f"  {key}. {value}\n"

        explanation_prompt += f"Correct Answer: {correct_ans}. {options.get(correct_ans, 'N/A')}\n"
        if not is_correct:
            explanation_prompt += f"Your Answer: {user_ans}. {options.get(user_ans, 'N/A')}\n"
            explanation_prompt += "Explain why the correct answer is correct and why the user's answer was incorrect, ONLY using the provided Context." # Emphasize context use
        else:
            explanation_prompt += "Explain why the correct answer is correct, ONLY using the provided Context." # Emphasize context use

        generated_explanation = "Error generating explanation." # Default error message
        try:
            # Use the 'model' object initialized in the previous step
            # 'model' is currently 'models/gemini-2.5-flash'
            explanation_response = model.generate_content(explanation_prompt)
            generated_explanation = explanation_response.text
        except Exception as e:
            generated_explanation = f"An error occurred while generating explanation with context: {e}"

        results.append({
            'question_index': q_index + 1,
            'question_text': question_text,
            'user_answer': user_ans,
            'correct_answer': correct_ans,
            'is_correct': is_correct,
            'explanation': generated_explanation
        })
    return score, len(parsed_quiz_data), results

print("'evaluate_quiz' function updated to include contextual explanation generation.")

# Call the updated evaluate_quiz function
total_score, total_questions, quiz_results_with_contextual_explanations = evaluate_quiz(user_quiz_responses, parsed_quiz, cleaned_text)

print("\n--- Quiz Results with Contextual Explanations ---")
print(f"You scored {total_score} out of {total_questions} questions correctly!")

for result in quiz_results_with_contextual_explanations:
    status = "Correct" if result['is_correct'] else "Incorrect"
    print(f"\nQuestion {result['question_index']}: {status}")
    print(f"  Question Text: {result['question_text']}")
    print(f"  Your Answer: {result['user_answer']}")
    print(f"  Correct Answer: {result['correct_answer']}")
    print(f"  Explanation: {result['explanation']}")
print("---------------------------------------------------")
```
